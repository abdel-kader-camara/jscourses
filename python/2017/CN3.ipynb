```python
from pandas.io.json import json_normalize
import pandas as pd
import json
import matplotlib.pyplot as plot
import datetime 

data = json.load(open('valeurs.json'))
dataframe = json_normalize(data["values"])
dataframe = dataframe.astype(dtype= {"Humidity" : "float32", "Lux" : "int32", 
                                     "Pression" : "int32", "Temp" : "float32",
                                      "Lux" : "int32", "date" : datetime.datetime})
plot.plot(dataframe["date"], dataframe["Humidity"])
plot.show()
```
